### Questão 1 a)

Observando o primeiro diagrama, para o caso sem atrito, as trações em todos os segumentos do cabo contínuo devem ser iguais. Analisando o ponto onde se aplica a força $F$ e realizando o equilíbrio estático naquele ponto, obtém-se:

$$
F = T_1
$$

Assim:

$$
F = T_1 = T_2 = T_3 = T_4 = T_5 = T_6
$$

Realizando o equilíbrio estático no DCL $(I)$:

$$
T_{10} = T_7 + T_6 = 2F \\
T_9 = T_5 + T_4 = 2F \\
T_8 = T_3 + T_2 = 2F
$$

Do DCL $(II)$:

$$
T_a + T_b = T_8 + T_9 + T_{10} = 6F
$$

Do DCL $(III)$:

$$
W = T_a + T_b = 6F
$$

Assim:

$$
F = \frac{W}{6} = 10\,\mathrm{kN}
$$

### Questão 1 b)

Analisando a subida do sistema, sabe-se que, da estática simples:

$$
\begin{array}{rcl}
W & = & T_a + T_b \\
  & = & T_8 + T_9 + T_{10} \\
  & = & T_2 + T_3 + T_4 + T_5 + T_6 + T_7\\ \hline
F & = & T_1
\end{array}
$$

Das relações de atrito:

$$
\begin{array}{ccc}
T_1 > T_2 & \Rightarrow & T_1 Q_- = T_2 \\ \hline
T_2 > T_3 & \Rightarrow & T_2 Q_- = T_3 \\ \hline
T_3 > T_4 & \Rightarrow & T_3 Q_- = T_4 \\ \hline
T_4 > T_5 & \Rightarrow & T_4 Q_- = T_5 \\ \hline
T_5 > T_6 & \Rightarrow & T_5 Q_- = T_6 \\ \hline
T_6 > T_7 & \Rightarrow & T_6 Q_- = T_7
\end{array}
$$

Assim, substituindo as relações:

$$
\begin{array}{rcl}
W & = & T_2 + T_3 + T_4 + T_5 + T_6 + T_7 \\
  & = & T_1 Q_- + T_2 Q_- + T_3 Q_- + T_4 Q_- + T_5 Q_- + T_6 Q_- \\
  & = & T_1 Q_- + T_1 Q_-^2 + T_2 Q_-^2 + T_3 Q_-^2 + T_4 Q_-^2 + T_5 Q_-^2 \\
  & = & T_1 Q_- + T_1 Q_-^2 + T_1 Q_-^3 + T_2 Q_-^3 + T_3 Q_-^3 + T_4 Q_-^3 \\
  & = & T_1 Q_- + T_1 Q_-^2 + T_1 Q_-^3 + T_1 Q_-^4 + T_2 Q_-^4 + T_3 Q_-^4 \\
  & = & T_1 Q_- + T_1 Q_-^2 + T_1 Q_-^3 + T_1 Q_-^4 + T_1 Q_-^5 + T_2 Q_-^5 \\
  & = & T_1 Q_- + T_1 Q_-^2 + T_1 Q_-^3 + T_1 Q_-^4 + T_1 Q_-^5 + T_1 Q_-^6 \\
  & = & T_1 \left(Q_- + Q_-^2 + Q_-^3 + Q_-^4 + Q_-^5 + Q_-^6\right) \\
  & = & F \left(Q_- + Q_-^2 + Q_-^3 + Q_-^4 + Q_-^5 + Q_-^6\right)
\end{array}
$$

Portanto:

$$
F = \frac{W}{Q_- + Q_-^2 + Q_-^3 + Q_-^4 + Q_-^5 + Q_-^6}
$$

Os resultados se encontram a seguir:

```python
W = 60      # kN
eps = 0.02  # ---

Qm = (1-eps)/(1+eps)

F = W/(Qm + Qm**2 + Qm**3 + Qm**4 + Qm**5 + Qm**6)

print(f'F = {F:.4f}  kN')
```

### Questão 1 c)

Para resolver este item, basta trocar $Q_-$ por $Q_+$ e refazer o cálculo. O resultado segue abaixo:

```python
Qp = 1/Qm

F = W/(Qp + Qp**2 + Qp**3 + Qp**4 + Qp**5 + Qp**6)

print(f'F = {F:.4f}  kN')
```

### Questão 1 d)

**Na situação de subida:**

Do DCL $(III)$, realizando o equilíbrio estático na vertical, obtém-se:

$$
T_a + T_b = W
$$

Do DCL $(II)$, podemos realizar a somatória dos momentos logo acima do ponto de aplicação da tração $T_a$ para obter:

$$
-T_{10} \frac{d}{2} + T_9 \frac{d}{2} + T_8 \frac{3d}{2} - T_b d = 0
$$

_(OBS: considerou-se os momentos no sentido horário como negativos por convenção.)_

Ou seja:

$$
T_b = \frac{3T_8 + T_9 - T_{10}}{2}
$$

Substituindo as relações anteriores da estática e do atrito:

$$
\begin{array}{rcl}
T_b & = & \displaystyle \frac{3(T_2+T_3) + T_4 + T_5 - T_6 - T_7}{2} \\
    & = & \displaystyle \frac{3(T_1 Q_- + T_2 Q_-) + T_3 Q_- + T_4 Q_- - T_5 Q_- - T_6 Q_-}{2} \\
    & = & \displaystyle \frac{3T_1(Q_- + Q_-^2) + T_2 Q_-^2 + T_3 Q_-^2 - T_4 Q_-^2 - T_5 Q_-^2}{2} \\
    & = & \displaystyle \frac{3T_1(Q_- + Q_-^2) + T_1 Q_-^3 + T_2 Q_-^3 - T_3 Q_-^3 - T_4 Q_-^3}{2} \\
    & = & \displaystyle \frac{3T_1(Q_- + Q_-^2) + T_1 Q_-^3 + T_1 Q_-^4 - T_2 Q_-^4 - T_3 Q_-^4}{2} \\
    & = & \displaystyle \frac{3T_1(Q_- + Q_-^2) + T_1 Q_-^3 + T_1 Q_-^4 - T_1 Q_-^5 - T_2 Q_-^5}{2} \\
    & = & \displaystyle T_1\left(\frac{3(Q_- + Q_-^2) + Q_-^3 + Q_-^4 - Q_-^5 - Q_-^6}{2}\right) \\
    & = & \displaystyle F\left(\frac{3(Q_- + Q_-^2) + Q_-^3 + Q_-^4 - Q_-^5 - Q_-^6}{2}\right) \\
    & = & \displaystyle W\left(\frac{3(Q_- + Q_-^2) + Q_-^3 + Q_-^4 - Q_-^5 - Q_-^6}{2(Q_- + Q_-^2 + Q_-^3 + Q_-^4 + Q_-^5 + Q_-^6)}\right) \\
\end{array}
$$

E finalmente:

$$
T_a = W - T_b
$$

**Na situação de descida**: da mesma forma, na situação de descida basta trocar $Q_-$ por $Q_+$.

Os resultados numéricos se encontram a seguir:

```python
Tb_sub = W*(3*(Qm + Qm**2) + Qm**3 + Qm**4 - Qm**5 - Qm**6) / \
         (2*(Qm + Qm**2 + Qm**3 + Qm**4 + Qm**5 + Qm**6))

Ta_sub = W - Tb_sub

print('SUBIDA:')
print(f'Ta = {Ta_sub:.4f} kN')
print(f'Tb = {Tb_sub:.4f} kN')
print('--------------------')

Tb_des = W*(3*(Qp + Qp**2) + Qp**3 + Qp**4 - Qp**5 - Qp**6) / \
         (2*(Qp + Qp**2 + Qp**3 + Qp**4 + Qp**5 + Qp**6))

Ta_des = W - Tb_des

print('DESCIDA:')
print(f'Ta = {Ta_des:.4f} kN')
print(f'Tb = {Tb_des:.4f} kN')
print('--------------------')
```

### Questão 3 b)

Analisando a fórmula para o perfil do vento, observa-se que $v_0 = V_r$ e $h_0 = 12\,\mathrm{m}$. Assim a fórmula para a força equivalente de vento fica na forma:

$$
F_{eq} = \frac{C_f \rho w V_r^2}{h_0^{2p}(2p+1)}\left(h_2^{2p+1}-h_1^{2p+1}\right)
$$

Como nesta questão $h_0=h_2$ e $h_1 = 0$, cabe a simplificação:

$$
F_{eq} = \frac{C_f \rho w h_2 V_r^2}{2p+1}
$$

Na área frontal $\varphi = 12\%$. Assim, das tabelas:

$$
C_f = \frac{2(4,13-5,18\varphi)}{3}
$$

O valor de $C_f$ e da força de vento equivalente são mostrados a seguir:

```python
%reset -f

phi_f = 0.12    # ---
rho = 1.2       # kg/m³
w_f = 3.5       # m
h2 = 12         # m
p = 0.25        # ---
Vr = 15         # m/s

Cf_f = 2*(4.13-5.18*phi_f)/3

Feq_f = Cf_f*rho*w_f*h2*(Vr**2)/(2*p+1)

print('Frontal:')
print(f'Cf = {Cf_f:.4f}')
print(f'Feq = {Feq_f:.4f} N')
```

### Questão 3 c)

Para calcular a condição limite de tombamento, é necessário fazer a somatória de momentos em torno do eixo de pivotamento da estrutura, que é o eixo que liga os pontos de apoio que ficam em contato com o solo durante o tombamento.

Neste caso, o braço da força de vento é $h_q$ e o braço da força peso é $2,75\,\mathrm{m}$. Assim:

$$
\sum M_p = -h_q F_{eq} + mg\cdot 2,75\,\mathrm{m} = 0
$$

Substituindo a relação para $F_{eq}$ e reorganizando:

$$
\frac{C_f \rho w h_2 V_r^2}{2p+1} = \frac{mg\cdot 2,75\,\mathrm{m}}{h_q}
$$

$$
\therefore \hspace{5mm} V_r = \sqrt{\frac{(2p+1)mg\cdot 2,75 \,\mathrm{m}}{C_f \rho w h_2 h_q}}
$$

O resultado numérico se encontra a seguir:

```python
from numpy import sqrt

m = 945   # kg
g = 9.81  # m/s²
hq = 7.2  # m

Vmax_f = sqrt(((2*p+1)*m*g*2.75)/(Cf_f*rho*w_f*h2*hq))

print(f'Para ventos frontais, Vr (max) = {Vmax_f:.4f} m/s')
```
