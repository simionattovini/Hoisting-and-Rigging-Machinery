**Subida**

- Do atrito:

Da polia $D$:
$$
T_1>T_2 \,\,\,\,\therefore\,\,\,\, T_1 Q_- = T_2
$$

Da polia $C$
$$
T_3 > T_4 \,\,\,\,\therefore\,\,\,\, T_3 Q_- = T_4
$$

Das polias $B$, $E$ e $F$:
$$
T_6 > T_7 > T_8 > T_9 \,\,\, \therefore \,\,\, \left\{\begin{array}{rcl}
T_6 Q_- & = & T_7 \\
T_7 Q_- & = & T_8 \\
T_8 Q_- & = & T_9
\end{array} \right.
$$

- Da estática:

No ponto $A$:
$$
F = T_1
$$

Na polia $D$:
$$
T_1 + T_2 = T_3
$$

Na polia $C$:
$$
T_5 = T_3 + T_4
$$

Na polia $B$:
$$
T_5 = T_6 + T_7 + T_8
$$

Na massa suspensa:
$$
W = T_2 + T_4 + T_9
$$

Partindo da equação da carga:

$$
W = T_2 + T_4 + T_9
$$

Substituições: $T_2 = T_1 Q_-$, $T_4 = T_3 Q_-$ e $T_9 = T_8 Q_-$

$$
W = T_1 Q_- + T_3Q_- + T_8 Q_-
$$

Substituições: $T_1 = F$, $T_3 = T_1 + T_2$, $T_8 = T_7 Q_-$ e $T_7 = T_6 Q_-$

$$
W = F Q_- + (T_1 + T_2)Q_- + T_6 Q_-^3
$$

Da estática, sabemos que $T_5 = T_6 + T_7 + T_8$. Das relações de atrito, sabemos que $T_7 = T_6 Q_-$ e $T_8 = T_6 Q_-^2$. Assim:

$$
T_5 = T_6 (1+Q_-+Q_-^2) \,\,\, \therefore \,\,\, T_6 = \frac{T_5}{1+Q_-+Q_-^2}
$$

Além disso, substituímos também $T_2 = T_1 Q_-$:

$$
W = F Q_- + T_1(1 + Q_-)Q_- + \left(\frac{T_5}{1+Q_-+Q_-^2}\right) Q_-^3
$$

Substituições: $T_1 = F$ e $T_5 = T_4 + T_3$:

$$
W = F(2 + Q_-)Q_- + \left(\frac{T_4+T_3}{1+Q_-+Q_-^2}\right) Q_-^3
$$

Substituição: $T_4 = T_3 Q_-$

$$
W = F(2 + Q_-)Q_- + \left(\frac{T_3(1+Q_-)}{1+Q_-+Q_-^2}\right) Q_-^3
$$

Substituição: $T_3 = T_1 + T_2$

$$
W = F(2 + Q_-)Q_- + \left(\frac{(T_2+T_1)(1+Q_-)}{1+Q_-+Q_-^2}\right) Q_-^3
$$

Substituição: $T_2 = T_1 Q_-$

$$
W = F(2 + Q_-)Q_- + \left(\frac{T_1(1+Q_-)^2}{1+Q_-+Q_-^2}\right) Q_-^3
$$

Substituição: $T_1 = F$

$$
W = F\left[(2 + Q_-)Q_- + \left(\frac{(1+Q_-)^2}{1+Q_-+Q_-^2}\right) Q_-^3\right]
$$

Reorganizando:

$$
\therefore \,\,\, F = \frac{W}{\displaystyle (2 + Q_-)Q_- + \left(\frac{(1+Q_-)^2}{1+Q_-+Q_-^2}\right) Q_-^3}
$$

### a)

```python
# a) Sem atrito

eps = 0               # adimensional
Qm = (1-eps)/(1+eps)  # adimensional
W = 300               # kgf

F = W/((2+Qm)*Qm + ((1+Qm)**2/(1+Qm+Qm**2))*Qm**3)

print(f'F = {F:.4f} kgf')
```

Na condição limite de equilíbrio, a normal na roda traseira é nula. Assim, podemos somar os momentos diretamente em torno do ponto de contato da roda dianteira com o solo:

$$
\sum M_A = -W_1 d+W_2\left((2,6+3+3) \cos 50^\circ +2,2 - d\right) = 0
$$

$$
\therefore \,\,\,\, d  = \frac{W_2 (8,6 \cos 50^\circ +2,2)}{W_1 + W_2}
$$

Os resultados seguem abaixo:

```python
# a)

%reset -f

from numpy import pi, cos

W1 = 800  # kgf
W2 = 300  # kgf

ang = 50*pi/180

d = W2*(8.6*cos(ang) + 2.2)/(W1 + W2)

print(f'd = {d:.4f} m')
```

### b)

Para que o sistema seja sempre estável, a roda traseira deve sempre gerar um momento positivo em torno de $A$. Por isso, desejamos que:

$$
\sum M_A < 0
$$

Do item anterior:
$$
\sum M_A = -W_1 d+W_2\left((2,6+3+3) \cos 50^\circ +2,2 - d\right) < 0
$$

O que resulta em:
$$
\therefore \,\,\,\, d  > \frac{W_2 (8,6 \cos 50^\circ +2,2)}{W_1 + W_2}
$$

Dividindo o lado direito no numerador e no denominador por $W_2$:

O que resulta em:
$$
\therefore \,\,\,\, d  > \frac{8,6 \cos 50^\circ +2,2}{\displaystyle\frac{W_1}{W_2} + 1}
$$

Se queremos que o sistema seja sempre estável, $W_1/W_2$ pode assumir qualquer valor não-negativo. Isso faz com que o pior caso para $d$ seja quando $W_1/W_2 = 0$. Portanto, a menor medida de $d$ que atende a este requisito é:

$$
d = 8,6 \cos 50^\circ +2,2
$$

```python
d = 8.6*cos(ang) + 2.2

print(f'd = {d:.4f} m')
```

## Questão 3 - Resolução

Considere o diagrama a seguir:

A tração $\vec{T}_{BA}$ e a tração $\vec{T}_{BC}$ são dadas pelas relações:

$$
\vec{T}_{BA} = T_{BA} \hat{e}_{BA}
$$

$$
\vec{T}_{BC} = T_{BC} \hat{e}_{BC}
$$

, sendo $\hat{e}_{BA}$ e $\hat{e}_{BC}$ os vetores unitários nas direções $BA$ e $BC$, respectivamente.

Os vetores braço são $\vec{r}_{DB}$ do ponto $D$ ao $B$ e $\vec{r}_{DW}$ ligando o ponto $D$ à porção mais alta do cabo que segura a carga.

A força relativa à carga é representada pelo vetor $\vec{W} = -800 \hat{k}$



### a)

A somatória dos momentos em torno de $D$ é dada por:

$$
\sum \vec{M}_D = \vec{r}_{DB} \times \left(\vec{T}_{BA} + \vec{T}_{BC} \right) + \vec{r}_{DW} \times \vec{W} = \vec{0}
$$

Reorganizando a equação, teremos:

$$
T_{BA}(\vec{r}_{DB} \times \hat{e}_{BA}) + T_{BC}(\vec{r}_{DB} \times \hat{e}_{BC})  = - \vec{r}_{DW} \times \vec{W}
$$

Os resultados dos produtos vetoriais são mostrados abaixo:



```python
%reset -f

import numpy as np

v_BA = np.asarray([2, -3, -6]).reshape((3,1))
e_BA = v_BA/np.linalg.norm(v_BA)

v_BC = np.asarray([-6, -3, -6]).reshape((3,1))
e_BC = v_BC/np.linalg.norm(v_BC)

r_DW = np.asarray([1, 4, 7.5]).reshape((3,1))

r_DB = np.asarray([0, 0, 6]).reshape((3,1))

W = np.asarray([0, 0, -800]).reshape((3,1))

m_BA = np.cross(r_DB.T, e_BA.T).T
m_BC = np.cross(r_DB.T, e_BC.T).T
m_W = np.cross(-r_DW.T, W.T).T

print(f'ê_BA = {e_BA.T}^T')
print(f'ê_BC = {e_BC.T}^T')
print(f'r_DB X ê_BA = {m_BA.T}^T')
print(f'r_DB X ê_BC = {m_BC.T}^T')
print(f'-r_DW X W = {m_W.T}^T')
```

Dos resultados, observa-se que na direção $z$, para a somatória dos momentos, se obtém $0=0$. Isso vem do fato de que todas as forças do sistema ou possuem uma reta base que passa pelo eixo $z$ ou são paralelas ao eixo $z$.

Então, resolve-se apenas as direções $\hat{i}$ e $\hat{j}$ da equação abaixo. Para isso, podemos definir:

$$
\vec{m}_{BA} = \vec{r}_{DB} \times \hat{e}_{BA}
\,\,\,\,\,\, | \,\,\,\,\,\,
\vec{m}_{BC} = \vec{r}_{DB} \times \hat{e}_{BC}
\,\,\,\,\,\, | \,\,\,\,\,\,
\vec{m}_{W} = -\vec{r}_{DW} \times \vec{W}
$$

Assim, podemos montar o seguinte sistema de equações:

$$
\left[
\begin{array}{cc}
m_{BAx} & m_{BCx} \\
m_{BAy} & m_{BCy}
\end{array}
\right]
\left\{
\begin{array}{c}
T_{BA} \\ T_{BC}
\end{array}
\right\}
=
\left\{
\begin{array}{c}
m_{Wx} \\ m_{Wy}
\end{array}
\right\}
$$

Para resolver o sistema basta inverter a matriz que multiplica as incógnitas. Assim:

$$
\left\{
\begin{array}{c}
T_{BA} \\ T_{BC}
\end{array}
\right\}
=
\left[
\begin{array}{cc}
m_{BAx} & m_{BCx} \\
m_{BAy} & m_{BCy}
\end{array}
\right]^{-1}
\left\{
\begin{array}{c}
m_{Wx} \\ m_{Wy}
\end{array}
\right\}
$$

Os resultados para as trações seguem:

```python
mA = np.zeros((2,2))

mA[0:2,0:1] = m_BA[0:2, 0:1]
mA[0:2,1:2] = m_BC[0:2, 0:1]

mB = m_W[0:2,0:1]

Tt = np.linalg.inv(mA) @ mB

T_BA = Tt[0,0]
T_BC = Tt[1,0]

print(f'T_BA = {T_BA:.2f} kgf')
print(f'T_BC = {T_BC:.2f} kgf')
```

### b)



Realizando a somatória das forças, temos:

$$
\sum \vec{F} = \vec{T}_{BA} + \vec{T}_{BC} + \vec{W} + R_x \hat{i} +R_y \hat{j} + R_z \hat{k} = \vec{0}
$$

Portanto:
$$
\left\{
\begin{array}{c}
R_x \\ R_y \\ R_z
\end{array}
\right\}
= - \vec{T}_{BA} - \vec{T}_{BC} - \vec{W}
$$

Os resultados seguem abaixo:

```python
vT_BA = T_BA*e_BA  # Reconstruindo o vetor de tração T_BA
vT_BC = T_BC*e_BC  # Reconstruindo o vetor de tração T_BC

R = - vT_BA - vT_BC - W

print(f'Rx = {R[0,0]:.2f} kgf')
print(f'Ry = {R[1,0]:.2f} kgf')
print(f'Rz = {R[2,0]:.2f} kgf')
```
