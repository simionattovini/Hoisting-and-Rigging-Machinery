### Questão 1 a)

No centro da placa, $y = 7,5\,\mathrm{m}$. Assim, calcula-se $v(y)$ na forma:

$$
v(7,5\,\mathrm{m}) = 0,501\,V_r\,(7,5)^{0,3} = \phi V_r
$$

O valor de $\phi$ é calculado a seguir:

```python
p = 0.3  # ---
phi = 0.501*(7.5**p)

print(f'phi = {phi:.4f}')
print(f'v(y) = {phi:.4f} * Vr')
```

### Questão 1 b)

Se a distribuição de pressão de vento é constante em toda a placa, a força equivalente deve ser aplicada exatamente no centróide de área da figura. Assim, no DCL mostrado acima, $h_q = 7,5\,\mathrm{m}$. Além disso a força de vento é dada por:

$$
F_{Eq} = \rho v^2(y)AC_f = 1,23 \left[\frac{\mathrm{kg}}{\mathrm{m^3}}\right]\left(\phi V_r\right)^2 \left[\frac{\mathrm{m}}{\mathrm{s}}\right]^2 (5\times 12)\left[\mathrm{m^2}\right] 1,98 = \phi'V_r^2 \,\,[\mathrm{N}]
$$

O valor de $\phi'$ é calculado a seguir:

```python
rho = 1.23  # kg/m³
hei = 5     # m
w = 12      # m
Cf = 1.98   # ---

phi_p = rho*(phi**2)*hei*w*Cf

print(f"phi' = {phi_p:.4f}")
print(f"F_eq = {phi_p:.4f} * Vr²")
```

Nesta situação, o contrapeso ainda não está instalado, então $P_2 = 0$. Estamos estudando a situação limite de equilíbrio, então $R_{Ay} = 0$. Já a força $P_1$ é dada por:

$$
P_1 = m_s\cdot g = 400\,\mathrm{kg} \cdot 9,81 \,\mathrm{m/s^2} = 3.924\,\mathrm{N}
$$

Assim, a somatória dos momentos em torno do ponto $B$, considerando o sentido anti-horário como positivo, resulta em:

$$
\sum M_B = P_1 \cdot 0,75\,\mathrm{m} - F_{Eq} h_q = 0
$$

Ou seja:

$$
F_{Eq} = \frac{P_1\cdot 0,75\,\mathrm{m}}{h_q} = \phi' V_r^2
$$

$$
\therefore \hspace{10mm} V_r = \sqrt{\frac{P_1\cdot 0,75\,\mathrm{m}}{h_q \phi'}}\,\,\, [\mathrm{m/s}]
$$

O resultado é mostrado a seguir:

```python
from numpy import sqrt

ms = 400   # kg
g = 9.81   # m/s²

P1 = ms*g  # N
hq1 = 7.5  # m

Vr1 = sqrt(P1*0.75/(hq1*phi_p))

print(f'Vr = {Vr1:.4f} m/s')
```

### Questão 1 c)

Neste caso a distribuição é variável ao longo de $y$. Do enunciado temos:

*   $p = 0,3$
*   $h_1 = 5 \,\mathrm{m}$
*   $h_2 = 10\,\mathrm{m}$

Assim, $h_q$ é dado por:

$$
h_q = \left(\frac{2p+1}{2p+2}\right)\left(\frac{h_2^{2p+2}-h_1^{2p+2}}{h_2^{2p+1}-h_1^{2p+1}}\right)
$$

O valor de $h_q$ é calculado a seguir:

```python
h1 = 5   # m
h2 = 10  # m

hq2 = ((2*p+1)/(2*p+2))*((h2**(2*p+2)-h1**(2*p+2))/(h2**(2*p+1)-h1**(2*p+1)))

print(f'hq = {hq2:.4f} m')
```

Já a força de vento é dada por:

$$
F_{Eq} = \frac{C_f \rho w k^2}{2p+1}\left(h_2^{2p+1}-h_1^{2p+1}\right)
$$

O único valor desta fórmula que ainda não conhecemos é o $k$. Porém, sabemos que:

$$
v(y) = ky^p = 0,501 V_r y^{0,3}
$$

Portanto, concluimos que $k = 0,501 V_r$. Assim, a fórmula para a força fica:

$$
F_{Eq} = \frac{C_f \rho w (0,501 V_r)^2}{2p+1}\left(h_2^{2p+1}-h_1^{2p+1}\right) = \phi'' V_r^2
$$

O valor de \phi'' é calculado a seguir:

```python
phi_pp = (Cf*rho*w*(0.501**2)/(2*p+1))*(h2**(2*p+1)-h1**(2*p+1))

print(f"phi'' = {phi_pp:.4f}")
print(f"F_eq = {phi_pp:.4f} * Vr²")
```

Neste caso, a somatória dos momentos resultará na mesma expressão anterior, mas com valor diferente para $h_q$ e $\phi''$ no lugar de $\phi'$. Assim, a velocidade máxima $V_r$ pode ser calculada por:

$$
V_r = \sqrt{\frac{P_1\cdot 0,75\,\mathrm{m}}{h_q \phi''}}\,\,\, [\mathrm{m/s}]
$$

Os resultados são mostrados a seguir:

```python
Vr2 = sqrt(P1*0.75/(hq2*phi_pp))

print(f'Vr = {Vr2:.4f} m/s')
```

### Questão 2 a)

Partindo da equação de movimento:

$$
m \ddot{y} + c \dot{y} + ky = cr \dot{\theta} + kr \theta
$$

Aplicamos a transformada de Laplace na equação com condições iniciais nulas. Assim:

$$
m s^2 Y(s) + c s Y(s) + k Y(s) = cr s\Theta(s) + kr \Theta(s)
$$

Reorganizando a equação, temos:

$$
\left(ms^2+cs+k\right)Y(s) = \left(crs+kr\right)\Theta(s)
$$

Uma vez que a saída é $y(t)$ e a entrada é $\theta(t)$, a função de transferência é dada por:

$$
H_1(s) = \frac{Y(s)}{\Theta(s)} = \frac{crs+kr}{ms²+cs+k}
$$

### Questão 2 b)

Partindo novamente da equação de movimento:

$$
m\ddot{y} + c\dot{y} + ky = cr \dot{\theta} + k\theta
$$

Aplica-se a transformada de Laplace com condições iniciais nulas:

$$
ms^2 Y(s) +cs Y(s) + kY(s) = crs\Theta(s) + kr\Theta(s)
$$

Reorganizando a equação, temos:

$$
ms^2 Y(s) = -cs\left(Y(s)-r\Theta(s)\right) - k \left(Y(s)-r\Theta(s)\right) = -(cs+k) \left(Y(s)-r\Theta(s)\right)
$$

A entrada solicitada pelo enunciado é $y(t) - r\theta(t)$. Assim, a função de transferência é dada por:

$$
H_2(s) = \frac{Y(s)}{Y(s)-r\Theta(s)} = \frac{-(cs+k)}{ms^2}
$$
